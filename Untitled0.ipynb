```python
!pip install -q keras-tcn --no-dependencies
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from tensorflow import keras
import tensorflow as tf

from tcn import TCN, tcn_full_summary
from sklearn.preprocessing import RobustScaler
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/gideonclottey/ReaseachProject1/main/stemdataset.csv', sep=',')
data
```

<h2><b>Data Preprocessing</h1>

```python
# The date and time are objects and hence we need to convert it
# Combine the date and Time
data['DateTime'] = data['Date'] + ' ' + data['Time']

#Move the combine date and time to the first comlumn and then delete the original date and time

first_column = data.pop('DateTime')
data.insert(0, 'DateTime', first_column)



# Coverting the DateTime  column into time stamp
data['DateTime'] = pd.to_datetime(data['DateTime'], format='%d/%m/%Y %H:%M')

data.head()
#Delete date and time column
data= data.drop(['Date', 'Time'], axis=1)
```

```python
data_input= data[['Air_Temperature', 'CO2 (ppm)', 'PAR', 'Relative_humidity_(%)', 'Tensiometer', 'Soil_Temperature','VPD_(kPa)',	'SDV_(mm)']]
data_input.shape[1]

data_input
```

```python
data_input.describe()
```

```python
color_list =[
    'blue',
    'orange',
    'green',
    'red',
    'purple',
    'cyan',
    'olive',
    'pink',
    'gray',
    'brown',
]

date_key ="DateTime"

def Visualize(data):
  time_data = data[date_key]
  features = list(data.select_dtypes(include=[np.number]).columns.values)
  feature_size = len(features)
  fig,axes = plt.subplots(
      nrows =int(np.ceil(feature_size/2)), ncols =2, figsize=(14, feature_size*2), dpi = 80, facecolor='w', edgecolor='k'
  )
  for i in range(feature_size):
    key =features[i]
    c = color_list[i % (len(color_list))]
    t_data = data[key]
    t_data.index =time_data
    t_data.head()
    ax = t_data.plot(
        ax = axes[i//2, i %2],
        color=c,
        title ="{}".format(key),
        rot=25
    )
    ax.legend([key])
    plt.tight_layout()

Visualize(data)
```

```python
RS = RobustScaler()
train_data_scaled = RS.fit_transform(data_input.values)
```

```python
n_features = train_data_scaled.shape[-1]
max_samples_per_sequence = 8

num_samples = train_data_scaled.shape[0]
num_sequences = num_samples // max_samples_per_sequence
remainder_samples = num_samples % max_samples_per_sequence
```

```python
train_data_reshaped = train_data_scaled[:num_sequences * max_samples_per_sequence]
train_data_reshaped = train_data_reshaped.reshape(-1, max_samples_per_sequence, n_features)
```

```python
targets = np.random.rand(num_sequences, 1)
```

```python
n_epochs = 50
n_splits = 5
```

<h1><b> Building Module</h1>

```python
from tensorflow.keras.utils import plot_model
# Create lists to store evaluation metrics
mae_scores = []
rmse_scores = []
mse_scores = []
r2_scores = []

kf = KFold(n_splits=n_splits, shuffle=False)

plt.figure(figsize=(14, 10))
for fold, (train_idx, test_idx) in enumerate(kf.split(train_data_reshaped, targets)):
    print('-'*15, '>', f'Fold {fold+1}', '<', '-'*15)
    X_train, X_valid = train_data_reshaped[train_idx], train_data_reshaped[test_idx]
    y_train, y_valid = targets[train_idx], targets[test_idx]

    scheduler = tf.keras.optimizers.schedules.ExponentialDecay(1e-3, 200*((len(train_data_reshaped)*0.8)/1024), 1e-5)

    model = keras.models.Sequential([
        keras.layers.LSTM(128, input_shape=(max_samples_per_sequence, n_features), return_sequences=True),
        TCN(input_shape=(max_samples_per_sequence, n_features), nb_filters=256, return_sequences=True, dilations=[1, 2, 4, 8, 16, 32]),
        keras.layers.Dense(1)
    ])

      # Print the model architecture
    print(model.summary())

    # Save the model architecture as a picture
    plot_model(model, to_file=f"model_architecture_fold_{fold+1}.png", show_shapes=True, show_layer_names=True)
    model.compile(optimizer="adam", loss="mae", metrics=keras.metrics.MeanAbsoluteError())

    history = model.fit(X_train, y_train,
                        validation_data=(X_valid, y_valid),
                        epochs=n_epochs,
                        batch_size=1024,
                        callbacks=[tf.keras.callbacks.LearningRateScheduler(scheduler)])

    # Evaluation on the validation data
    y_pred = model.predict(X_valid).squeeze().mean(axis=1)  # Average predictions over the second axis
    mae = np.mean(np.abs(y_valid.squeeze() - y_pred))
    mse = np.mean((y_valid.squeeze() - y_pred) ** 2)
    rmse = np.sqrt(np.mean((y_valid.squeeze() - y_pred) ** 2))
    r2 = 1 - (np.sum((y_valid.squeeze() - y_pred) ** 2) / np.sum((y_valid.squeeze() - np.mean(y_valid.squeeze())) ** 2))

    mae_scores.append(mae)
    rmse_scores.append(rmse)
    mse_scores.append(mse)
    r2_scores.append(r2)

    model.save(f'Fold{fold+1}_weights')

# Create a DataFrame to store the evaluation metrics
metrics_df = pd.DataFrame({'Fold': range(1, n_splits + 1), 'MAE': mae_scores, 'RMSE': rmse_scores, 'MSE': mse_scores, 'R^2': r2_scores})

# Print the evaluation metrics in tabular form
print(metrics_df)

# Plot the evaluation metrics
plt.figure(figsize=(10, 6))

# Plot MAE and RMSE on the same graph
plt.plot(metrics_df['Fold'], metrics_df['MAE'], marker='o', label='MAE')
plt.plot(metrics_df['Fold'], metrics_df['RMSE'], marker='o', color='orange', label='RMSE')
plt.plot(metrics_df['Fold'], metrics_df['MSE'], marker='o', color='green', label='MSE')
plt.plot(metrics_df['Fold'], metrics_df['R^2'], marker='o', color='red', label='R^2')


plt.title('Evaluation Metrics Across Folds')
plt.xlabel('Fold')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```

<h3><b/>LSTM Model<h3/>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Create lists to store evaluation metrics for LSTM
mae_scores_lstm = []
rmse_scores_lstm = []
mse_scores_lstm = []
r2_scores_lstm = []


kf = KFold(n_splits=n_splits, shuffle=False)

plt.figure(figsize=(14, 10))

# Loop through each fold
for fold, (train_idx, test_idx) in enumerate(kf.split(train_data_reshaped, targets)):
    print('-'*15, '>', f'Fold {fold+1}', '<', '-'*15)
    X_train, X_valid = train_data_reshaped[train_idx], train_data_reshaped[test_idx]
    y_train, y_valid = targets[train_idx], targets[test_idx]

    scheduler = tf.keras.optimizers.schedules.ExponentialDecay(1e-3, 200*((len(train_data_reshaped)*0.8)/1024), 1e-5)

    model = keras.models.Sequential([
        LSTM(128, input_shape=(max_samples_per_sequence, n_features), return_sequences=False),  # Changed to return_sequences=False
        Dense(1)
    ])

    # Print the model architecture
    print(model.summary())

    model.compile(optimizer="adam", loss="mae", metrics=[keras.metrics.MeanAbsoluteError()])

    history = model.fit(X_train, y_train,
                        validation_data=(X_valid, y_valid),
                        epochs=n_epochs,
                        batch_size=1024,
                        callbacks=[tf.keras.callbacks.LearningRateScheduler(scheduler)])

    # Make predictions using the LSTM model
    y_pred_lstm = model.predict(X_valid).squeeze()

    # Calculate evaluation metrics for LSTM
    mae_lstm = mean_absolute_error(y_valid, y_pred_lstm)
    mse_lstm = mean_squared_error(y_valid, y_pred_lstm)
    rmse_lstm = np.sqrt(mse_lstm)
    r2_lstm = r2_score(y_valid, y_pred_lstm)

    mae_scores_lstm.append(mae_lstm)
    rmse_scores_lstm.append(rmse_lstm)
    mse_scores_lstm.append(mse_lstm)
    r2_scores_lstm.append(r2_lstm)

    model.save(f'LSTM_Fold{fold+1}_weights')

# Create a DataFrame to store the evaluation metrics for LSTM
metrics_df_lstm = pd.DataFrame({'Fold': range(1, n_splits + 1),
                                'MAE_LSTM': mae_scores_lstm,
                                'RMSE_LSTM': rmse_scores_lstm,
                                'MSE_LSTM': mse_scores_lstm,
                                'R^2_LSTM': r2_scores_lstm})

# Print the evaluation metrics for LSTM in tabular form
print(metrics_df_lstm)

# Plot the evaluation metrics for LSTM
plt.figure(figsize=(10, 6))

# Plot MAE, RMSE, MSE, and R^2 for LSTM on the same graph
plt.plot(metrics_df_lstm['Fold'], metrics_df_lstm['MAE_LSTM'], marker='o', label='MAE LSTM')
plt.plot(metrics_df_lstm['Fold'], metrics_df_lstm['RMSE_LSTM'], marker='o', color='orange', label='RMSE LSTM')
plt.plot(metrics_df_lstm['Fold'], metrics_df_lstm['MSE_LSTM'], marker='o', color='green', label='MSE LSTM')
plt.plot(metrics_df_lstm['Fold'], metrics_df_lstm['R^2_LSTM'], marker='o', color='red', label='R^2 LSTM')

plt.title('LSTM Evaluation Metrics Across Folds')
plt.xlabel('Fold')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create lists to store evaluation metrics
mae_scores_rf = []
rmse_scores_rf = []
mse_scores_rf = []
r2_scores_rf = []

# Initialize K-fold cross-validation
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=False)

plt.figure(figsize=(14, 10))

# Ensure train_data_reshaped and targets have the same number of samples
print("train_data_reshaped shape:", train_data_reshaped.shape)
print("targets shape:", targets.shape)

# Check if the dimensions of train_data_reshaped and targets match
if train_data_reshaped.shape[0] != targets.shape[0]:
    print("Mismatch in the number of samples between train_data_reshaped and targets.")

    # Assuming train_data_reshaped is (samples, time_steps, features) and targets is (samples,)
    # Correct the mismatch by aligning the data or removing inconsistent samples
    if train_data_reshaped.shape[0] > targets.shape[0]:
        train_data_reshaped = train_data_reshaped[:targets.shape[0], :, :]
    else:
        targets = targets[:train_data_reshaped.shape[0]]

    print("Data corrected. Now train_data_reshaped and targets have the same number of samples.")

# Proceed with K-fold cross-validation
# Proceed with K-fold cross-validation
for fold, (train_idx, test_idx) in enumerate(kf.split(train_data_reshaped, targets)):
    print('-'*15, '>', f'Fold {fold+1}', '<', '-'*15)
    X_train, X_valid = train_data_reshaped[train_idx], train_data_reshaped[test_idx]
    y_train, y_valid = targets[train_idx], targets[test_idx]

    # Flatten the 3D time series data into a 2D array
    X_train_flattened = X_train.reshape(X_train.shape[0], -1)

    # Define and train the Random Forest model
    model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
    model_rf.fit(X_train_flattened, y_train)

    # Flatten the 3D time series validation data into a 2D array
    X_valid_flattened = X_valid.reshape(X_valid.shape[0], -1)

    # Make predictions using the Random Forest model
    y_pred_rf = model_rf.predict(X_valid_flattened)

    # ... rest of the loop code ...

    # Calculate evaluation metrics
    mae_rf = mean_absolute_error(y_valid, y_pred_rf)
    mse_rf = mean_squared_error(y_valid, y_pred_rf)
    rmse_rf = np.sqrt(mse_rf)
    r2_rf = r2_score(y_valid, y_pred_rf)

    mae_scores_rf.append(mae_rf)
    rmse_scores_rf.append(rmse_rf)
    mse_scores_rf.append(mse_rf)
    r2_scores_rf.append(r2_rf)

# Create a DataFrame to store the evaluation metrics for Random Forest
metrics_df_rf = pd.DataFrame({'Fold': range(1, n_splits + 1), 'MAE_RF': mae_scores_rf, 'RMSE_RF': rmse_scores_rf, 'MSE_RF': mse_scores_rf, 'R^2_RF': r2_scores_rf})

# Print the evaluation metrics for Random Forest in tabular form
print(metrics_df_rf)

# Plot the evaluation metrics for LSTM, TCN, and Random Forest
plt.figure(figsize=(10, 6))

# Plot MAE and RMSE for all models on the same graph
plt.plot(metrics_df['Fold'], metrics_df['MAE'], marker='o', label='MAE LSTM')
plt.plot(metrics_df['Fold'], metrics_df['RMSE'], marker='o', color='orange', label='RMSE LSTM')
plt.plot(metrics_df_rf['Fold'], metrics_df_rf['MAE_RF'], marker='o', label='MAE RF')
plt.plot(metrics_df_rf['Fold'], metrics_df_rf['RMSE_RF'], marker='o', color='green', label='RMSE RF')

plt.title('Comparison of Evaluation Metrics Across Folds')
plt.xlabel('Fold')
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.show()
```
